### Perguntas a serem respondidas a partir dos dados.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ler arquivo csv para um datafram
development = pd.read_csv('./3.1-data-sheet-udemy-courses-web-development.csv')
desing = pd.read_csv('./3.1-data-sheet-udemy-courses-design-courses.csv')
music = pd.read_csv('./3.1-data-sheet-udemy-courses-music-courses.csv')
business = pd.read_csv('./3.1-data-sheet-udemy-courses-business-courses.csv')

```

1. Quantos cursos estão disponíveis na plataforma?

```python
todos_cursos = pd.concat([development, desing, music, business])

todos_cursos.info()


print('\nO total de cursos disponíveis na plataforma é:',todos_cursos['course_id'].value_counts().sum())

```

2. Quais e quantos são os cursos que abordam o assunto: JavaScript?

```python
todos_cursos["course_title"] = todos_cursos["course_title"].str.lower()

js = todos_cursos[todos_cursos['course_title'].str.contains('javascript', na= False)]
totalJavascript = len(js)
js_cursos = js[['course_title']]


print(f'Nossa plataforma disponibiliza de cursos sendo {totalJavascript} cursos com JavaScript. Nossa lista de cursos disponíveis: ',todos_cursos['course_id'].value_counts().sum())
```

3. Qual é o preço médio (coluna *price*) dos cursos oferecidos na plataforma?

```python
mean_todos_cursos = todos_cursos['price'].mean()
print('O preço médio na coluna price é:  R$', round(mean_todos_cursos,2))
```

4. Quais são os cinco cursos com maior número de inscritos (coluna *num_subscribers*)?

```python
inscritos = todos_cursos.sort_values(by='num_subscribers',ascending=False).dropna().head()
inscritos
```

5. Dos cinco cursos mais populares em termos de inscritos, mostre o **rate médio**, **máximo** e **mínimo** de cada um deles.

```python
max_inscritos = inscritos['Rating'].max()
min_inscritos = inscritos['Rating'].min()
med_inscritos  = inscritos['Rating'].mean()


print(f'O maximo de inscritos: {max_inscritos}, o mínimo de inscritos foi: {min_inscritos} e a média de inscritos é: {round(med_inscritos,2)}' )
```

6. Apresente os dez cursos mais visualizados na plataforma.

```python
todos_sort = todos_cursos.sort_values(by='num_reviews',ascending=False)
todos_sort.dropna().head(10)
```

7. A partir dos dez cursos mais vistos, mostre: 1) qual tem o maior número de inscritos; 2) qual tem o maior **rate** (avaliação do curso).

```python
mais_vistos = todos_cursos.sort_values(by='num_reviews',ascending=False)
mais_vistos.dropna().head(10)

vistos = mais_vistos['num_subscribers'].max()
m_vistos =  mais_vistos['Rating'].max()

print(f'Dos mais vistos o maior número de inscritos: {vistos} e com maior avaliação do curso foi: {m_vistos}')
```

8. Dos cursos listados na base de dados, qual tem maior duração em horas?

```python
maximo_horas = todos_cursos.groupby(by=['content_duration','course_title']).sum().sort_values(by='content_duration', ascending=False).head(5)
maximo_horas
```

9. Dos cursos listados na base de dados, qual tem o maior número de aulas (*lectures*)?

```python
maximo_lectures = todos_cursos.groupby(by=['num_lectures','course_title']).sum().sort_values(by='num_lectures', ascending=False)
maximo_lectures.head(5)

```

10. Apresente o número (contagem) de cursos agrupados por nível (coluna *level*).

```python
gp_nivel = todos_cursos.groupby("level")["course_title"].count().plot(kind='bar',figsize=(10,7))

```

11. Quais são os cursos mais recentes contidos na base de dados?

```python
recente =  todos_cursos.sort_values('published_timestamp', ascending=False)
recente.dropna().head(10)
```

12. Apresente o número (contagem) de cursos agrupados por nível (coluna *level*) e por assunto (coluna *subject*).

```python
gp_cont = todos_cursos.groupby(["level", "subject"])["course_title"].count().plot(kind='bar',figsize=(10,7),colormap='prism')
```
